```python
import csv

csv_data = [['Name', 'Age', 'Experience'],['Oswaldo', 22, 10]]

with open('./student_data.csv', 'r') as file:
    whole_file = csv.reader(file)
    for row in whole_file:
        print(row)

with open('./write_csv.csv', 'w') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerows(csv_data)
```

```python
import json 

json_data = {'name': 'Oswaldo', 'Age': 22, 'Ocupation': 'Analytics Engineer', 'Experience': 4}

with open('./student_data.json', 'r') as file:
    data = json.load(file)
    print(data)

with open('./write_json.json', 'w') as file:
    json.dump(json_data, file)
    
```
